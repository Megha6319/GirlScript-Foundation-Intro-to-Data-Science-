```python
#DATA VISUALIZATION IS USED WHEN WE WANT THE DATA TO BE VISUALIZED IN THE FORM OF GRAPH AND PIE CHART
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
chocolates=[1,2,3,4,5,6,7]
max_t=[39,45,66,77,98,99,87]
min_t=[40,41,44,43,47,41,44]
avg_t=[50,51,49,52,54,46,50]
```

```python
plt.xlabel('Biscuits')
plt.ylabel('CHIPS')
plt.title('SNACKKSS')

plt.plot(chocolates,max_t,label='Max')
plt.plot(chocolates,min_t,label='Min')
plt.plot(chocolates,avg_t,label='Avg')
#The attribute Loc in legend() is used to specify the location of the legend. Default value of loc is loc=”best” (upper left). The strings 'upper left', 'u
plt.legend(loc='best')
```

```python
plt.plot(chocolates,max_t,label='Max')
plt.plot(chocolates,min_t,label='Min')
plt.plot(chocolates,avg_t,label='Avg')

#setting location of legend by specifing number
plt.legend(loc=5)
```

```python
exp_values=[1400,600,300]
exp_name=['mom','dad','sis']
```

```python
plt.pie(exp_values,labels=exp_name)
plt.show()
```

```python
# to show axis equally
plt.axis('equal')

plt.pie(exp_values,labels=exp_name)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5, autopct='%0.1f%%')
plt.show()
```
