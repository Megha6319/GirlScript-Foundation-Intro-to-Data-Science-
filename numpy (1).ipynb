```python
#joining arrays
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr = np.concatenate((arr1,arr2),axis=0)
print(arr)
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
arr = np.concatenate((arr1, arr2), axis=1)

print(arr)
```

```python
#Stacking
arr1=np.array([2,4,6])
arr2=np.array([8,6,4])
arr = np.stack((arr1,arr2), axis =1)    
print(arr)
```

```python
#HorizontalStacking
arr1=np.array([2,4,6])
arr2=np.array([8,6,4])
arr = np.hstack((arr1,arr2))
print(arr)
```

```python
#verticalStacking
arr1=np.array([2,4,6])
arr2=np.array([8,6,4])
arr = np.vstack((arr1,arr2))
print(arr)
```
