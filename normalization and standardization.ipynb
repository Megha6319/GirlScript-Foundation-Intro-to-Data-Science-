 Normalization AND Standardization 

```python
#NORMALIZATION:-IT IS A SCALING TECHNIQUE IN WHICH VALUES ARE SHUFTED AND RESCALED SO THAT THEY END UP RANGIN BETWEEN 0 AND 1.
#Xnorm= (X - Xmin) / (Xmax - Xmin)  [FORMULA TO CALCULATE NORMALIZATION]
```

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler


from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('titanic.csv')
data.head()
```

```python
data.describe()
```

```python
scaling=MinMaxScaler()
```

```python
#WE FIND NORMALIZATION HERE BY SCALIN THE LARGER VALUES TO 0 AND 1
scaling.fit_transform(data[['PassengerId','Fare']])
```

```python
# let's look at missing data

data.isnull().sum()
```

```python
#TEST AND TRAIN THE DATA TO FIND THE ACCURACY IB THE DATASET,MORE THE ACCURACY BETTER THE MODEL
X_train,X_test,y_train,y_test=train_test_split(data[['PassengerId','Fare']], data.Survived, test_size=0.3,
                                                random_state=0)
X_train.shape, X_test.shape
```

```python
#STANDARDIZATION:Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation.
#formula to calculate standardization:z = (x - x_mean) / std
```

```python

scaler = StandardScaler() # create an object
X_train_scaled = scaler.fit_transform(X_train) # fit the scaler to the train set, and then transform it
X_test_scaled = scaler.transform(X_test)
```

```python
print('means (Pclass, Age and Fare): ', X_train_scaled.mean(axis=0))
print('std (Pclass, Age and Fare): ', X_train_scaled.std(axis=0))
```
